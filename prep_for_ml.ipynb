```python
def price_diff(s1, s2):
  s3 = []
  for i in range(len(s1)):
    if (s2[i] - s1[i] >= 1):
      s3.append(1)
    elif (s2[i] - s1[i] <= -1):
      s3.append(-1) 
    else:
      s3.append(0)
  return s3
```

```python
Co30_d = price_diff(big_df['Close'], big_df['CloseOver30']) 
```

```python
big_df['Co30_d'] = Co30_d
```

```python
Co60_d = price_diff(big_df['Close'], big_df['CloseOver60']) 
```

```python
big_df['Co60_d'] = Co60_d
```

```python
Co90_d = price_diff(big_df['Close'], big_df['CloseOver90']) 
```

```python
big_df['Co90_d'] = Co90_d
```

```python
def vol_diff(s1, s2):
  s3 = []
  for i in range(len(s1)):
    if (s2[i] - s1[i] >= 100):
      s3.append(1)
    elif (s2[i] - s1[i] <= -100):
      s3.append(-1) 
    else:
      s3.append(0)
  return s3
```

```python
Vo30_d = price_diff(big_df['Volume'], big_df['VolumeOver30']) 
```

```python
big_df['Vo30_d'] = Vo30_d
```

```python
Vo60_d = price_diff(big_df['Volume'], big_df['VolumeOver60']) 
```

```python
big_df['Vo60_d'] = Vo60_d
```

```python
Vo90_d = price_diff(big_df['Volume'], big_df['VolumeOver90']) 
```

```python
big_df['Vo90_d'] = Vo90_d
```

```python
# import dependencies
import pandas as pd 
import matplotlib.pyplot as plt
import random
import datetime
import re
# import nltk
```

### Generating Dummy Data

```python
"""
Generating dummy data to test classifiers

Preliminary schema:
[Date, Subreddit, Username, Ticker, Text, Sentiment]

"""
# Generating subreddits
sub_list = ['wallstreetbets', 'stocks', 'investing', 'robinhood', 'stockmarket', 'securityanalysis', 'daytrading']
subreddits = [sub_list[random.randrange(len(sub_list)-1)] for i in range(50)]
subreddits
```

```python
# Generating usernames
usernames = ["user" + str(i) for i in range(50)]
usernames
```

```python
# Generating sentences
tickers = ['AAPL', 'GOOG', 'FB', 'NFLX', 'MSFT', 'AMZN']
sent = ['is going down', 
        'is going up', 
        'to the moon', 
        'to 0', 
        'is amazing value', 
        'is overpriced',
        'is undervalued']
text = ['$' + tickers[random.randrange(len(tickers)-1)] + ' ' + sent[random.randrange(len(sent)-1)] for i in range(50)]
text
```

```python
tickers = [re.search('\$[A-Z]+', t)[0] for t in text]
tickers
```

```python
# Generate sentiment
sentiment = [round(random.random()) for i in range(50)]
sentiment
```

```python
# Generating dates
start_date = datetime.date(2020, 1, 1)
end_date = datetime.date(2020, 12, 31)

time_between_dates = end_date - start_date
days_between_dates = time_between_dates.days

dates = [start_date + datetime.timedelta(days=random.randrange(days_between_dates)) for i in range(50)]
dates = [d.strftime("%Y-%m-%d") for d in dates]
dates
```

```python
# create preliminary array of tuples to pass to dictionary
dummy_tup = zip(dates, subreddits, usernames, tickers, text, sentiment)
dummy_list = list(dummy_tup)
dummy_list
```

```python
# generate list of dictionaries from tuple
dummy_data = [{'date': i[0],
               'subreddit': i[1],
               'username': i[2],
               'ticker': i[3],
               'text': i[4],
               'sentiment': i[5]}
               for i in dummy_list]
dummy_data  
```

```python
# Convert to dataframe
dummy_df = pd.DataFrame(dummy_data)
dummy_df.head()
```

```python
dummy_df.to_csv("dummy_nlp_data.csv")
```

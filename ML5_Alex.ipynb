```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import tensorflow as tf

df = pd.read_csv('cleaned_sentinment.csv')
df = df.drop(['Unnamed: 0','date','ticker'],axis=1)
df.head()
```

```python
df.columns
```

```python
df = df.drop(['CloseOver30','VolumeOver30','CloseOver60','VolumeOver60','CloseOver90','VolumeOver90','Close','Volume'],axis=1)
df.head()
```

```python
df['overall_sent_comp']=pd.to_numeric(df['overall_sent_comp'],errors='coerce')
```

```python
for column in df.columns:
  if pd.api.types.infer_dtype(df[column]) == 'mixed':
    df[column]=pd.to_numeric(df[column],errors='coerce')
```

```python
df = df.dropna()
```

```python
for column in df.columns:
  print(column + ' ' + pd.api.types.infer_dtype(df[column]))
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(df['verb_tense'].values.reshape(-1,1)))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(['verb_tense'])
encode_df.head()
```

```python
clean_df =df.merge(encode_df,left_index=True,right_index=True).drop(['verb_tense'],axis=1)
clean_df.head()
```

```python
y=clean_df['Co30_d'].values
X=clean_df.drop(['Co30_d','Co60_d','Co90_d','Vo30_d','Vo60_d','Vo90_d'],axis=1).values

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

```python
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 40
#hidden_nodes_layer1 = 20

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1,input_dim=number_input_features,activation='relu'))


#nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1,input_dim=number_input_features,activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="mae", optimizer="Adadelta", metrics=["accuracy"])
```

```python
fit_model = nn.fit(X_train,y_train,epochs=1)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
